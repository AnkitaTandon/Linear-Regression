```python
drop_list=['PassengerId','Name','Gender','Ticket']
```

```python
df[drop_list]
```

```python
df2=df.drop(drop_list,axis=1)
```

```python
df2.head()
```

```python
x=df2.drop('Survived',axis=1)
y=df2['Survived']
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
 lr=lr.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
x=lr.predict(x_train)
```

```python
acc=accuracy_score(y_train,x)
ps=precision_score(y_train,x)
recall=recall_score(y_train,x)
f1=f1_score(y_train,x)
```

```python
print('Accuracy Score =',acc*100)
```

```python
print('Precision Score =',ps*100)
```

```python
print('Recall Score =',recall*100)
```

```python
print('F1 Score =',f1*100)
```

```python
x=lr.predict(x_test)
acc=accuracy_score(y_test,x)
ps=precision_score(y_test,x)
recall=recall_score(y_test,x)
f1=f1_score(y_test,x)
```

```python
print('Accuracy Score =',acc*100)
print('Precision Score =',ps*100)
print('Recall Score =',recall*100)
print('F1 Score =',f1*100)
```
